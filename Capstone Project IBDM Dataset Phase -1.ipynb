```python
# Total number of titles to scrape
total_titles = 1751
```

```python
# construct counters for paging
items_per_page = 50
pages = total_titles // items_per_page
if total_titles % items_per_page != 0:
    pages += 1
```

```python
# creating required lists
Sno = []
Movie_Name = []
Director_1=[]
Director_2=[]
Director_3=[]
Duration = []
Year = []
Ratings = []
Metascore = []

Stars_1=[]
Stars_2=[]
Stars_3=[]
Stars_4=[]
Votes=[]
Genre_1=[]
Genre_2=[]
Genre_3=[]
Gross=[]
Certification=[]

# Extracting movie details from each page
for page in range(pages):
    # Find all the movie elements
    movie_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]')

    # Iterating through the elements and extract the details
    for element in movie_elements:
        # Extracting Serial number
        try:
            sno_element = element.find_element(By.XPATH, './/span[@class="lister-item-index unbold text-primary"]')
            Sno.append(sno_element.text.replace('.', ''))
        except NoSuchElementException:
            Sno.append('Not Available')

        # Extracting Movie Name
        try:
            movie_name_element = element.find_element(By.XPATH, './/h3[@class="lister-item-header"]/a')
            Movie_Name.append(movie_name_element.text)
        except NoSuchElementException:
            Movie_Name.append('Not Available')
        
        
        #try:
            #director_element = element.find_element(By.XPATH,'.//p[@class="text-muted"]/span[@class="ghost"]/following-sibling::a') #'.//p[@class="text-muted"]/span[1]')
            #Director_Name.append(director_element.text)
        #except NoSuchElementException:
            #Director_Name.append('Not Available')

        # Extracting Duration
        try:
            duration_element = element.find_element(By.XPATH, './/span[@class="runtime"]')
            Duration.append(duration_element.text)
        except NoSuchElementException:
            Duration.append('Not Available')

        # Extracting Year
        try:
            year_element = element.find_element(By.XPATH, './/span[@class="lister-item-year text-muted unbold"]')
            Year.append(year_element.text.strip('()'))
        except NoSuchElementException:
            Year.append('Not Available')

        # Extracting Ratings
        try:
            ratings_element = element.find_element(By.XPATH, './/div[@class="inline-block ratings-imdb-rating"]/strong')
            Ratings.append(ratings_element.text)
        except NoSuchElementException:
            Ratings.append('Not Available')

        # Extracting Metascore
        try:
            metascore_element = element.find_element(By.XPATH, './/div[@class="inline-block ratings-metascore"]/span')
            Metascore.append(metascore_element.text.strip())
        except NoSuchElementException:
            Metascore.append('Not Available')
            

    # Extracting Director Name
    # To extract director names with sub-fields.
    dire=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]')
    for i in dire:
        director_text = i.text
        directors = re.findall(r"Director[s]?:\s+(.*)", director_text)
        if directors:             
            Director_1.append(directors[-1].split("|")[0].split(",")[0])
            try:
                Director_2.append(directors[-1].split("|")[0].split(",")[1])
            except:
                Director_2.append("Not Available")
            try:
                Director_3.append(directors[-1].split("|")[0].split(",")[2])
            except:
                Director_3.append("Not Available") 
        
            
    # To extracting votes and gross collection
    vote=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in vote:
        Votes.append(i.text.split(":")[1].split(" ")[1])
    for i in vote:
        try:
            Gross.append(i.text.split("|")[1].split("$")[1].strip("M"))
        except:
            Gross.append("Not Available")
                
    # To extracting certification
    cer=driver.find_elements(By.XPATH,'//p[@class="text-muted "]')
    for i in cer:
        try:
            (i.text.split("|")[2]) # To verify extraction condition
            Certification.append(i.text.split("|")[0])
        except:
            Certification.append("Not Available")
                
    # To extracting genres with sub-fields
    gen=driver.find_elements(By.XPATH,'//span[@class="genre"]')
    for i in gen:
        Genre_1.append(i.text.split(",")[0])
        try:
            Genre_2.append(i.text.split(",")[1])
        except:
            Genre_2.append("Not Available")
        try:
            Genre_3.append(i.text.split(",")[2])
        except:
            Genre_3.append("Not Available")
                
    # To extracting stars names with sub-fields
    star=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]')
    for i in dire:
        director_text = i.text
        directors = re.findall(r"Director[s]?:\s+(.*)", director_text)
        if directors:    
            Stars_1.append(directors[-1].split(":")[1].split(",")[0])
            try:
                Stars_2.append(directors[-1].split(":")[1].split(",")[1])
            except:
                Stars_2.append("Not Available")
            try:
                Stars_3.append(directors[-1].split(":")[1].split(",")[2])
            except:
                Stars_3.append("Not Available")
            try:
                Stars_4.append(directors[-1].split(":")[1].split(",")[3])
            except:
                Stars_4.append("Not Available")
        
```
